```python
import pandas as pd
```

```python
# import dataset 
spotify_df = pd.read_csv('dataset.csv')
```

```python
# take a look at dataset
spotify_df.head()
```

```python
# check to see if there are any null values
spotify_df.info()
```

```python
# verify that only 'artists', 'album_name', and 'track_name' have null values
nan_count = spotify_df.isna().sum()
print(nan_count)
```

```python
# remove rows with null values. Since only one row has null values, it is unlikely to make a difference in final model. 
spotify_df.dropna(inplace=True)
spotify_df.info()
```

```python
spotify_df.describe()
```

```python
# check what data is in 'Unnamed: 0' column
spotify_df.iloc[:,:1]
```

```python
# remove 'Unnamed: 0'
spotify_df = spotify_df.iloc[: , 1:]
spotify_df.head()
```

```python
spotify_df.describe()
```

```python
# change miliseconds to seconds
spotify_df['duration_minutes'] = spotify_df['duration_ms'] / 1000 / 60
spotify_df.describe()

# subset of spotify_df without 'duration_ms'
new_col = [col for col in spotify_df.columns if col != 'duration_ms']
df = spotify_df[new_col]

df.head()
```
