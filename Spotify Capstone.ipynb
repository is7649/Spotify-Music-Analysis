```python
# import necesary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Wrangling

```python
df = pd.read_csv('/Users/itssel/Desktop/Spotify_Dataset.csv')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
# get rid of unnamed column
df = df.iloc[: , 1:]
df.head()
```

```python
# drop any duplicate rows
df2 = df.drop_duplicates(keep = 'first')
display(df2)
```

All duplicate rows have been dropped using drop_duplicates.

```python
cols = []
for i in df2.columns:
    cols.append(i)
print(cols)
```

```python
# need to make new df to see where there are duplicates
# group by includes track_genre
column_list = ['track_id', 'artists', 'album_name', 'track_name', 'popularity', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'track_genre', 'duration_minutes']
groupby_list = ['track_id', 'artists', 'track_genre','track_name', 'album_name']
df_groupby = df2[column_list].groupby(groupby_list).count().sort_values(by= 'danceability', ascending = False )
display(df_groupby)
```

```python
groupby_list2 = ['track_id', 'artists','track_name', 'album_name']
df_groupby2 = df2[column_list].groupby(groupby_list2).count().sort_values(by= 'popularity', ascending = False )
display(df_groupby2)
```

It appears that the track_genre is responsible for many duplicates in this dataset. According to Spotify's Web API, there is no specific genre for each track. There are only genres for artists. Due to this, I've made decision to remove the track_genre column. 

```python
df3 = df2.drop(columns= ['track_genre'])
df3 = df3.drop_duplicates(keep = 'first')
display(df3)
```

Explicit column will be converted from True/False to 1/0 respectively. 

```python
df3["explicit"] = df3["explicit"].astype(int)
display(df3)
```

Exploratory Data Analysis

```python
import warnings
warnings.simplefilter("ignore", UserWarning)
sns.distplot(df3['popularity']).set_title('Popularity Distribution')
```

There appears to be a huge amount of tracks with a popularity of 0. Another noteworthy observation is that there is a right skewness to the data, which lets us know that high popularity scores are rare.

```python
# visualize distribution for numeric columns (float or int)
# these are the numeric columns
numeric_cols = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'duration_minutes', 'explicit']
# histograms
for column in numeric_cols:
    plt.hist(df3[column])
    plt.title("Distribution of " + column.capitalize())
    plt.ylabel("Count")
    plt.xlabel(column.capitalize())
    plt.show()
```

It appears that the danceability, tempo, and perhaps valence have a normal distribution. The distribution for duration in minutes appears to be skewed, but this may be due to the scaling of the x-axis. Time signature is skewed to the left, but this makes sense because 4/4 is the most commonly used time signature in music. The distribution of energy looks like it is increasing as energy increases. The distribution of key does not have a clear pattern. Liveness, instrumentalness, acousticness, and speechiness are right skewed. The distributions of mode and explicit are bimodal. The distribution of loudness is left skewed.

Comparing features to one another

```python
plt.subplots(figsize=(12,12))
sns.heatmap(df3[numeric_cols].corr(), annot = True);
```

By examining this correlation matrix, we can tell that there is a high correlation between energy & loudness (r=0.76) and energy & acousticness (r=-0.73). Loudness & acousticness (r = -0.58), valence & danceability (0.49), and instrumentalness & loudness (r=-0.43) are moderately correlated. The remaining pairs have little to no correlation between one another. 

I chose the six features with the highest correlations to use later during feature engineering. The features are as follows: energy, loudness, acousticness, valence, danceability, and instrumentalness.

PRE PROCESSING 

```python
# key features (highest correlations)
features = ['energy', 'loudness', 'acousticness', 'valence', 'danceability', 'instrumentalness']
X = df3[features]
y = df3['popularity']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X[features], y, train_size=0.5, random_state=40)
```

```python
X_train.head()
```

```python
scaler = StandardScaler().fit(X_train)
```

```python
scaler.transform(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
print(X_train_scaled.mean(axis=0), X_train_scaled.std(axis=0))
```
