Step 1: Exploring each individual variable

```python
import warnings
warnings.simplefilter("ignore", UserWarning)
sns.distplot(df['popularity']).set_title('Popularity Distribution')
```

```python
# visualize distribution for numeric columns (float or int)
# these are the numeric columns
numeric_cols = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'duration_minutes']
# histograms
for column in numeric_cols:
    plt.hist(df[column])
    plt.title("Distribution of " + column.capitalize())
    plt.ylabel("Count")
    plt.xlabel(column.capitalize())
    plt.show()
```

It appears that the danceability, tempo, and perhaps valence have a normal distribution. The distribution for duration in minutes appears to be skewed, but this may be due to the scaling of the x-axis. Time signature is skewed to the left, but this makes sense because 4/4 is the most commonly used time signature in music. The distribution of energy looks like it is increasing as energy increases. The distribution of key does not have a clear pattern. Liveness, acousticness, and speechiness are right skewed. The distribution of mode is bimodal. The distribution of instrumentalness is unimodal.

Step 2: Explore Relationships between variables and Popularity Index

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16,4))

for xcol, ax in zip(['danceability', 'energy', 'key', 'loudness'], axes):
    df.plot(kind='scatter', x=xcol, y='popularity', ax=ax, alpha=0.006, color='purple')
```

The data is very large. As a result, it appears that I am overplotting. To try to make the scatterplots more useful, I've adjusted the alpha value for the graphs above. I am also going to try to take a random sample of 1000 to try and see a clearer trend.

Comparing categorical features to popularity

```python
#Calculate the average 'popularity' score by genre
track_genre_avg_pop = df.groupby('track_genre')['popularity'].mean()
track_genre_avg_pop.head()
```

```python
track_genre_avg_pop.hist(bins=30)
plt.title('Distribution of track genre average popularity')
plt.xlabel('Average popularity')
plt.ylabel('count');
```
