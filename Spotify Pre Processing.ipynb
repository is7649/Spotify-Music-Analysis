Comparing features to one another

```python
plt.subplots(figsize=(12,12))
heatmap_cols = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'duration_minutes', 'explicit', 'popularity']
sns.heatmap(df3[heatmap_cols].corr(), annot = True);
```

By examining this correlation matrix, we can tell that there is a high correlation between energy & loudness (r=0.76) and energy & acousticness (r=-0.72). Loudness & acousticness (r = -0.57), valence & danceability (0.48), and instrumentalness & loudness (r=-0.43) are moderately correlated. The remaining pairs have little to no correlation between one another. 

I chose the four features with the highest correlations to use later during feature engineering. The features are as follows: instrumentalness, speechiness, danceability, and loudness. By squaring our highest correlation coefficient, R, we get the coefficient of determination (R²) that we need to clear: .04.

PRE PROCESSING 

```python
# key features (highest correlations)
features = ['speechiness', 'loudness', 'danceability', 'instrumentalness']
X = df3[features]
y = df3['popularity']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X[features], y, train_size=0.5, random_state=40)
```

```python
X_train.head()
```

```python
scaler = StandardScaler().fit(X_train)
```

```python
scaler.mean_
```

```python
scaler.scale_
```

```python
scaler.transform(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
print(X_train_scaled.mean(axis=0), X_train_scaled.std(axis=0))
```

```python
scaler = StandardScaler().fit(X_test)
X_test_scaled = scaler.transform(X_test)
```

```python
print(X_test_scaled.mean(axis=0), X_test_scaled.std(axis=0))
```

MODELING

Model 1: Linear Regression (four key features)

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lm = lr.fit(X_train_scaled, y_train)

lm_r_sq = round((lm.score(X_test_scaled, y_test) *100), 2)
print(f'The coefficient of determination is: {lm_r_sq}%')
```

The R-Squared for this linear model is approximately 5.33%, which is extremely low.

Assess Linear Model Performance

```python
y_te_pred = lm.predict(X_test_scaled)
```

```python
from sklearn import metrics
lm_MAE = round(metrics.mean_absolute_error(y_test, y_te_pred),2)
lm_MSE = round(metrics.mean_squared_error(y_test, y_te_pred), 2)
lm_RMSE = round(np.sqrt(metrics.mean_squared_error(y_test, y_te_pred)),2)
print(f'MAE: {lm_MAE}')
print(f'MSE: {lm_MSE}')
print(f'RMSE: {lm_RMSE}')
```

An average of 17.68 spread for the residuals in the prediction model for a range of 0–100 in ‘popularity’ is not good. This indicates that our predictions are far from our actual popularity values. 

Linear Model (all features)

For this next model, I will repeat the linear regression but instead will include all features.

```python
X2 = df3[numeric_cols]
X2_train, X2_test, y2_train, y2_test = train_test_split(X2[numeric_cols], y, train_size=0.5, random_state=40)
```

```python
scaler = StandardScaler().fit(X2_train)
X2_train_scaled = scaler.transform(X2_train)
```

```python
scaler = StandardScaler().fit(X2_test)
X2_test_scaled = scaler.transform(X2_test)
```

```python
lr = LinearRegression()
lm = lr.fit(X2_train_scaled, y2_train)

lm_r_sq = round((lm.score(X2_test_scaled, y2_test) *100), 2)
print(f'The coefficient of determination is: {lm_r_sq}%')
```

```python
y2_te_pred = lm.predict(X2_test_scaled)
```

```python
lm2_MAE = round(metrics.mean_absolute_error(y2_test, y2_te_pred),2)
lm2_MSE = round(metrics.mean_squared_error(y2_test, y2_te_pred), 2)
lm2_RMSE = round(np.sqrt(metrics.mean_squared_error(y2_test, y2_te_pred)),2)
print(f'MAE: {lm2_MAE}')
print(f'MSE: {lm2_MSE}')
print(f'RMSE: {lm2_RMSE}')
```

An average of 17.48 spread for the residuals in the prediction model for a range of 0–100 in ‘popularity’ is still not good. However, there is a slight improvement in RMSE and R2 when using all features instead of just the key four.
