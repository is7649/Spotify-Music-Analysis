DECISION TREE MODEL

```python
from sklearn.tree import DecisionTreeRegressor
# we are using all features 
X3 = df3[all_features]
y3 = df3['track_popularity']

X3_train, X3_test, y3_train, y3_test=train_test_split(X3, y3, 
                                                  test_size = 0.25,
                                                  random_state = 246)
```

```python
max_depth_range = range(1, 10)
RMSE_scores = []
from sklearn.model_selection import cross_val_score, cross_validate, GridSearchCV, learning_curve
for depth in max_depth_range:
    treereg = DecisionTreeRegressor(max_depth=depth, random_state=1)
    MSE_scores = cross_val_score(treereg, X3, y3, cv=5, scoring='neg_mean_squared_error')
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))
plt.plot(max_depth_range, RMSE_scores);
plt.xlabel('max_depth');
plt.ylabel('RMSE');
```

It appears that a max_depth of 9 is optimum because it shows the lowest RMSE of approximately 12.5, which is worse than the linear model's RMSE. It also has a fairly big max depth, which may lead to overfitting. 

```python
from sklearn import tree
gini_model = tree.DecisionTreeClassifier(criterion='gini', random_state = 1234, max_depth = 9)
gini_model.fit(X3_train, y3_train)
y3_pred = gini_model.predict(X3_test)
y3_pred = pd.Series(y3_pred)
gini_model
```
